```python
import csv
import os
import subprocess
from string import punctuation

import pandas as pd

import spacy
import spacy_transformers
from wordfreq import zipf_frequency
```

```python
def count_csv_elements_in_file(filepath):
    total_elements = 0
    with open(filepath, "r", encoding="utf-8") as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            total_elements += len(row)
    
    return total_elements

language = []
total_words = []

for path, subdirs, files in os.walk("raw-word-lists"):
    for name in files:
        if name.endswith(".csv"):   # ðŸ‘ˆ important
            filename = os.path.join(path, name)
            language.append(name.split(".")[0])
            total_words.append(count_csv_elements_in_file(filename))
        
```
